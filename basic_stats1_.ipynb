```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sales_data_with_discounts.csv')

categorical_cols = df.select_dtypes(include='object').columns.tolist()
if 'Date' in categorical_cols:
    categorical_cols.remove('Date')

print(" Bar Charts: Total Net Sales Value by Category ")

for col in categorical_cols:
    plt.figure(figsize=(8, 5))

    category_sales = df.groupby(col)['Net Sales Value'].sum().sort_values(ascending=False)

    # Convert Net Sales Value by dividing by 100
    category_sales_divided_by_100 = category_sales / 100

    # bar plot
    sns.barplot(x=category_sales_divided_by_100.index, y=category_sales_divided_by_100.values, palette='viridis')

    plt.title(f'Total Net Sales Value by {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Total Net Sales Value (Divided by 100)', fontsize=12)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

print("--- Bar Chart Findings: Top Revenue Contributors ---")
print("Day: Peak revenue day is Thursday, indicating when sales are most financially impactful.")
print("SKU: The 'Cash cow' products, bringing in the most money, are M07, M10 & M09.")
print("BU: The primary financial driver is Mobiles.")
print("Brand: The top revenue-contributing brand is Orange.")
print("Model: The highest-earning specific product models are s-20 and o-11.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sales_data_with_discounts.csv')

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

print("--- Concept of Standardization (Z-score Normalization) ---")
print("""
Standardization (Z-score Normalization) is a data scaling technique that
transforms data to have a mean of 0 and a standard deviation of 1.

Formula: z = (x - μ) / σ
Why use it?
1.  **Equal Contribution:** It ensures that all features contribute equally
    to the distance calculations in many machine learning algorithms (e.g.,
    K-Means, SVMs, PCA, Logistic Regression, Neural Networks). Features
    with larger ranges don't disproportionately influence the model.
""")

print("\n--- Data Distribution BEFORE Standardization ---")
#histograms for numerical columns BEFORE standardization
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30, label=f'Original {col}')
    plt.title(f'Original Distribution of {col}', fontsize=12)
    plt.xlabel(col, fontsize=10)
    plt.ylabel('Frequency', fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=8)
    plt.tight_layout()
    plt.show()

print("\n--- Performing Standardization ---")

df_numerical_standardized = df[numerical_cols].copy()

# Standardize each numerical column manually
for col in numerical_cols:
    mean_val = df_numerical_standardized[col].mean()
    std_val = df_numerical_standardized[col].std()

    # Avoid division by zero if standard deviation is 0
    if std_val == 0:
        df_numerical_standardized[col] = 0
        print(f"Warning: Standard deviation for '{col}' is 0. Setting standardized values to 0.")
    else:
        df_numerical_standardized[col] = (df_numerical_standardized[col] - mean_val) / std_val
    print(f"Standardized column: {col}")

print("\n--- Data Distribution AFTER Standardization ---")
#histograms for numerical columns AFTER standardization
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df_numerical_standardized[col], kde=True, bins=30, color='orange', label=f'Standardized {col}')
    plt.title(f'Standardized Distribution of {col}', fontsize=12)
    plt.xlabel(f'{col} (Standardized Z-score)', fontsize=10)
    plt.ylabel('Frequency', fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=8)
    plt.tight_layout()
    plt.show()

print("\n--- Before and After Comparison Insights ---")
print("""
**Observations from Before and After Histograms:**

1.  **Shape Preservation:** The shape of the distribution for each column (e.g., skewness) remains the same. Standardization only scales and shifts the data, it does not change its fundamental shape.
2.  **Centered at Zero:** After standardization, the distribution of each numerical column is centered around 0. This is evident as the bulk of the data (and the peak of the KDE curve) will be near 0 on the x-axis.
3.  **Scaled Range:** The range of values on the x-axis for each column has changed. Instead of the original large or varied ranges, the values are now typically within a smaller range, often between -3 and +3, reflecting units of standard deviation from the mean.
4.  **Uniform Scale:** All numerical columns are now on a comparable scale, which is beneficial for algorithms sensitive to feature magnitudes.

This transformation ensures that no single feature dominates the learning process purely because of its larger numerical range.
""")
```

```python
import pandas as pd

df = pd.read_csv('sales_data_with_discounts.csv')

print("--- Need for Converting Categorical Data into Dummy Variables (One-Hot Encoding) ---")
print("""
Machine Learning algorithms, especially most traditional ones (like Linear Regression,
Logistic Regression, Support Vector Machines, K-Nearest Neighbors, etc.),
are designed to work with numerical input data. They perform mathematical
operations (e.g., calculating distances, gradients, coefficients) that
are not meaningful with text-based or categorical labels directly.

**Why One-Hot Encoding?**
1.  **Numerical Representation:** It converts categorical variables into a numerical format
    that algorithms can understand and process.
2.  **Avoid Ordinality Assumption:** If categories were simply mapped to numbers (e.g., Red=1, Green=2, Blue=3),
    the algorithm might mistakenly assume an ordinal relationship (e.g., Green is "greater" than Red),
    which is often not true for nominal categories. One-hot encoding avoids this by
    creating separate binary columns for each category.
3.  **No Magnitude Implication:** Each new binary column (dummy variable) represents the
    presence (1) or absence (0) of a specific category, without implying any
    magnitude or order between category""")

# Identify categorical columns
# Select object dtype. 'Date' is typically handled separately or dropped for ML,
# but for demonstration of one-hot encoding, we'll include it if it's an object type.
categorical_cols = df.select_dtypes(include='object').columns.tolist()

print(f"\n--- Original Categorical Columns Identified: {categorical_cols} ---")
print("\nFirst 5 rows of original DataFrame (showing categorical columns):")
print(df[categorical_cols].head())

print("\n--- Applying One-Hot Encoding to Categorical Columns ---")

# Apply one-hot encoding
# drop_first=True is often used to avoid multicollinearity in regression models,
# but for general transformation display, we'll keep all dummies for clarity.
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=False)

print("\n--- Displaying a Portion of the Transformed Dataset (First 5 Rows) ---")
# Display the first few rows of the transformed DataFrame
display_cols = list(df.select_dtypes(exclude='object').columns) + \
               [col for col in df_encoded.columns if col.startswith(tuple(categorical_cols))][:10] # Show first 10 dummy cols

print(df_encoded[display_cols].head())

print("\n--- Shape of Original vs. Transformed Dataset ---")
print(f"Original DataFrame shape: {df.shape}")
print(f"Transformed DataFrame shape: {df_encoded.shape}")
print("\nNotice the increase in the number of columns due to one-hot encoding.")
```
